# Epic 2: Visualizing and Analysing the Data

Covers: Import and Read Dataset, Univariate Analysis, Bivariate Analysis, Multivariate Analysis

## Story: Import and Read Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Epic 1 - Data Collection and Architecture Design/data/loan_data.csv')
df.head()
```

## Story: Univariate Analysis

Analyse individual columns (distribution of income, loan amount, categorical counts, etc.)

```python
df['ApplicantIncome'].hist(bins=30)
plt.title('Applicant Income Distribution')
plt.show()
```

## Story: Bivariate Analysis

Analyse relationship between two variables (e.g. Education vs Loan_Status)

```python
sns.countplot(x='Education', hue='Loan_Status', data=df)
plt.show()
```

## Story: Multivariate Analysis

Analyse relationships between multiple variables (e.g. correlation heatmap)

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```
